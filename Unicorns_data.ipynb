### Missing Values

```python
print('missing values (%) per column: \n', 100*data.isnull().mean())
```

```python
#locate rows with missing values 
missing_vals = data[data['Investors'].isnull()]
missing_vals
```

It appears that the data has been mis-aligned in these particular rows. The values which seem to be missing are for the City column.



```python
missing_vals.columns
```

```python
print(missing_vals.columns.tolist()) 
```

```python
#lets duplicate the city column to shift the data from one column to another 
missing_vals['Duplicate Industry'] = missing_vals['Industry']
missing_vals
```

```python
#columns to shift 
condition = missing_vals['Investors'].isnull()

missing_vals.loc[condition, 'Industry'] = missing_vals.loc[condition, 'City']
missing_vals.loc[condition, 'Investors'] = missing_vals.loc[condition, 'Duplicate Industry']
```

```python
pd.set_option('display.max_colwidth', None)
missing_vals.loc[10]
```

```python
#drop duplicate col in missing_vals df

missing_vals = missing_vals.drop(columns = ['Duplicate Industry'])
missing_vals
```

 Now, let's fill the city column with the mode corresponding to the country column. 

```python
#mode city value by country 
mode_city = data.groupby('Country')['City'].apply(lambda x : x.mode()[0]).to_frame().reset_index()
mode_city
```

```python
missing_vals['Country'].unique()
```

The mode exercise has provided a useful value for China, however, there seems to be an error with the other three countries. For simplicity, let's fill the city value with the country value.  

```python
# function to fill incorrect values in city column with mode value

def fill_city_col(value):
    if value == 'Bahamas':
        return 'Bahamas'
    elif value == 'Singapore':
        return 'Singapore'
    elif value == 'Hong Kong':
        return 'Hong Kong'
    elif value == 'China':
        return 'Beijing'
    else:
        pass 
            
```

```python
missing_vals['City'] = missing_vals['Country'].apply(fill_city_col)
```

```python
#check df
missing_vals
```

```python
#There seems to be an error on row 917, let's fix this. 
missing_vals.loc[917]
```

```python
missing_vals.at[917, 'Industry'] = 'Mobile & telecommunications'
```

```python
mode_investor = data.groupby('Country')['Investors'].apply(lambda x : x.mode()).to_frame().reset_index()
mode_investor
```

```python
mode_investor.loc[mode_investor['Country']=='China']
```

### Reshape Data

For better analysis, let's split the investors column, drop the dollar symbol in the valuation column and change the data type to float.

```python
#split the investors column by the comma delimiter 
data[['Investor1', 'Investor2', 'Investor3', 'Investor4']] = data['Investors'].str.split(',', expand=True)
```

```python
data = data.drop(columns='Investors')
```

```python
data['Valuation ($B)'] = data['Valuation ($B)'].str.replace('$', '')
```

```python
data.head()
```

```python
data['Valuation ($B)'] = data['Valuation ($B)'].astype(float)
```

```python
data.dtypes
```

```python
#let's move on to splitting the date column into month and year

data['Date Joined'] = pd.to_datetime(data['Date Joined'])

data['Month'] = data['Date Joined'].dt.month
data['Year'] = data['Date Joined'].dt.year
```

```python
data.head()
```

```python
#for further analysis, let's summarise the industries by number of investors 
ind_invdf = data.groupby('Industry')[['Investor1','Investor2','Investor3','Investor4']].count()
ind_invdf
```

```python
ind_invdf['Total Number of Investors'] = ind_invdf[['Investor1','Investor2','Investor3','Investor4']].sum(axis=1)
ind_invdf.drop(columns=['Investor1','Investor2','Investor3','Investor4'])
```

```python
#sum the number of unique investors to give the total number of companies invested in by investor
investor_cols = data[['Investor1', 'Investor2', 'Investor3', 'Investor4']]
combined_investors = pd.concat([investor_cols[col] for col in investor_cols.columns]).dropna()
```

```python
combined_investors
```

```python
unique_count = combined_investors.value_counts()
unique_count = unique_count.to_frame().reset_index()
```

```python
unique_count.rename(columns={'index':'Investor', 'count':'Number of Companies Invested'})
```

## Data Storage 

```python
data.columns
```

```python
data = data.rename(columns = {'Valuation ($B)': 'Valuation_billion_dollars', 'Date Joined' : 'Date_Joined'})
data.head()
```

```python
data.dtypes
```

```python
#before loading data into sqlite db, let's change some of the data types 

data['Valuation_billion_dollars'] = data['Valuation_billion_dollars'].astype('float')
data['Date_Joined'] = data['Date_Joined'].astype('str')
```

```python
#now all missing values have been dealt with let's save the normalized data in a csv format and we can crack on with the analysis!
data.to_csv('unicorn_clean_data.csv', index = False)
unique_count.to_csv('unicorn_investor_data.csv', index = False)
```

```python
#let's also insert and save the data to an sqlite database

import sqlite3


conn = sqlite3.connect('unicorn.db')

cursor = conn.cursor()

create_table_query = '''
CREATE TABLE IF NOT EXISTS Unicorns (
Company_ID INTEGER PRIMARY KEY, 
Company TEXT,
Valuation_billion_dollars REAL, 
Date_Joined TEXT, 
Country TEXT,
City TEXT,
Industry TEXT,
Investor1 TEXT,
Investor2 TEXT,
Investor3 TEXT,
Investor4 TEXT,
Month INT, 
Year INT
)
'''

cursor.execute(create_table_query)

insert_query = '''
INSERT INTO Unicorns (Company, Valuation_billion_dollars, Date_Joined, Country, City, Industry, Investor1, Investor2, Investor3, Investor4, Month, Year) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?);
'''

for index, row in data.iterrows():
    cursor.execute(insert_query, (row['Company'], row['Valuation_billion_dollars'], row['Date_Joined'], row['Country'], row['City'], row['Industry'], row['Investor1'], row['Investor2'], row['Investor3'], row['Investor4'], row['Month'], row['Year']))
    
conn.commit()
conn.close()

print('Data inserted successfully into SQLite DB.')
```
